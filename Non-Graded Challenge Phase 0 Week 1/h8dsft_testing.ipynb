# Task

```python
class Book: # Nama class nya adalah Book
  '''
  Class Book ditujukan untuk sebuah objek buku yang ada di dalam catalog,
  dimana objek tersebut memiliki tiga buah attribute, yaitu judul buku
  (title), penulis buku (author), dan ID unik dari buku (unique_id).
  '''

  # Attribute
  def __init__(self, title, author, unique_id):
    '''
    Fungsi __init__ digunakan untuk mendefinisikan attribute. Pada fungsi
    __init__ terdapat tiga parameter, yaitu title, author, dan unique_id.
    '''

    self.title = title # Attribute title memiliki value dari parameter title
    self.author = author # Attribute author memiliki value dari parameter author
    self.unique_id = unique_id # Attribute unique_id memiliki value dari parameter unique_id
```

```python
class Catalog: # Nama class nya adalah Catalog
  '''
  Class Catalog ditujukan untuk sebuah objek katalog buku, dimana objek
  tersebut memiliki sebuah attribute, yaitu books.

  Attribute books adalah sebuah dictionary yang isinya adalah key-value
  dari objek buku pada class Book.

  Class ini juga memiliki beberapa method, yaitu menambahkan buku ke
  katalog, mencari buku dari nama penulisnya, menghapus buku dari katalog,
  melihat isi dari katalog, dan menjalankan user interface menu.
  '''

  # Attribute
  def __init__(self):
    '''
    Fungsi __init__ digunakan untuk mendefinisikan attribute.
    '''

    self.books = {}
    # Attribute books adalah dictionary yang kosong (inisialisasi dictionary)
    # Menyimpan data key-value dari semua buku yang sudah ditambahkan ke dalam katalog

  # Method
  def add_book(self, book_title, book_author, book_id):
    '''
    Fungsi add_book berfungsi untuk menambahkan buku ke dalam katalog. Fungsi
    ini memiliki tiga parameter, yaitu book_title, book_author, dan book_id.
    '''

    check_book = self.books.get(book_id) # Mengecek apakah buku dengan ID uniknya sudah ada di dalam katalog
    if check_book is not None: # Jika ID buku tersebut ditemukan di dalam katalog
      raise ValueError("Buku dengan ID tersebut sudah ada di dalam katalog.") # Beri tahu user bahwa buku dengan ID tersebut sudah ada di dalam katalog
    else: # Jika buku dengan ID tersebut tidak ada di dalam katalog
      self.books[book_id] = Book(book_title, book_author, book_id) # Menambahkan buku ke dalam dictionary books (katalog buku)
      # Key dari dictionary books adalah book_id dan setiap buku memiliki tiga buah value, yaitu book_title, book_author, dan book_id

  def find_book(self, search_book_author):
    '''
    Fungsi find_book berfungsi untuk mencari buku di dalam katalog, berdasarkan
    nama authornya. Fungsi ini memiliki sebuah parameter, yaitu search_book_author.
    '''

    book_found = [] # Membuat list book_found untuk mengumpulkan buku dengan nama penulis yang sedang dicari
    for book in self.books.values(): # For loop untuk memeriksa setiap buku yang ada di dalam katalog
      if search_book_author.lower() in book.author.lower(): # Jika nama author buku sama dengan nama author yang sedang dicari
        book_found.append(book) # menambahkan buku tersebut kedalam list book_found
    return book_found # mereturn list book_found yang berisi buku-buku dengan nama penulisnya adalah nama penulis yang sedang dicari

  def remove_book(self, book_id):
    '''
    Fungsi remove_book berfungsi untuk menghapus buku dari katalog. Fungsi ini
    memiliki sebuah parameter, yaitu book_id.
    '''

    if book_id not in self.books: # Jika ID buku tidak ada di dalam katalog
      raise ValueError("Buku dengan ID tersebut tidak ada di dalam katalog.") # Beri tahu user bahwa buku dengan ID tersebut tidak ada di dalam katalog
    else: # Jika buku dengan ID tersebut ada di dalam katalog
      del self.books[book_id] # Menghapus buku dari dalam dictionary books (katalog buku)

  def show_book(self):
    '''
    Fungsi show_boook berfungsi untuk menampilkan semua buku yang ada di dalam katalog.
    '''

    return list(self.books.values()) # Mereturn list dari katalog untuk menampilkan semua buku yang ada di dalam katalog

  def run_user_interface(self):
    '''
    Fungsi run_user_interface berfungsi untuk menjalankan user interface menu.
    Di dalam menu, terdapat beberapa pilihan, yang dapat dipilih oleh user.

    Isi dari menu adalah menambahkan buku ke dalam katalog, mencari buku berdasarkan
    nama penulis, menghapus buku dari katalog, menampilkan isi dari katalog, dan
    keluar dari program.
    '''

    print("Selamat Datang di Katalog Buku\n") # Greeting kepada user

    while True: # While loop untuk menampilkan menu pada user interface
      print("\nMenu:") # Pilihan menu
      print("1. Menambah Buku") # Pilihan pertama, yaitu method nambah buku ke dalam katalog
      print("2. Mencari Buku Berdasarkan Nama Penulis") # Pilihan kedua, yaitu method mencari buku berdasarkan nama penulis
      print("3. Hapus Buku") # Pilihan ketiga, yaitu method menghapus buku dari katalog
      print("4. Tampilkan Buku di Katalog") # Pilihan keempat, menampilkan semua buku yang ada di dalam katalog
      print("5. Exit\n") # Pilihan kelima, untuk keluar dari program
      pilih = input("Pilihan: ") # Meminta user input untuk menu

      if pilih == '1': # Jika user memilih 1
        title = input("Judul Buku: ") # Meminta user menginput judul buku
        author = input("Penulis Buku: ") # Meminta user menginput nama penulis dari buku
        unique_id = input("ID Unik dari Buku: ") # Meminta user menginput ID buku
        self.add_book(title, author, unique_id) # Menambahkan buku ke dalam katalog
        print("Buku berhasil ditambahkan ke dalam katalog.\n") # Memberi tahu user bahwa buku telah ditambahkan ke dalam katalog

      elif pilih == '2': # Jika user memilih 2
        search_author = input("Nama Penulis yang Dicari: ") # Meminta user menginput nama penulis yang ingin dicari
        found_book = self.find_book(search_author) # Memanggil fungsi find_book dengan parameter value nya adalah search_author
        if len(found_book) > 0: # Jika terdapat buku dengan penulis yaitu penulis yang dicari
          print("Buku yang ditemukan:") # Memberi tahu user bahwa buku dengan nama penulis yang dicari telah ditemukan
          for book in found_book: # For loop untuk mencetak detail buku-buku yang telah ditemukan
            print(f"Judul: {book.title}, Penulis: {book.author}, ID Buku: {book.unique_id}") # Mencetak detail setiap buku dari buku-buku yang telah ditemukan
        else: # Jika tidak terdapat buku dengan nama penulis yang dicari
          print("Buku dengan nama penulis tersebut tidak ditemukan.\n") # Memberi tahu user bahwa tidak ada buku dengan nama penulis yang dicari

      elif pilih == '3': # Jika user memilih 3
        book_id = input("ID Buku yang Akan Dihapus: ") # Meminta user menginput ID dari buku
        try: # Try except untuk mencegah error
          self.remove_book(book_id) # Menghapus buku dari katalog
          print("Buku berhasil dihapus dari katalog.\n") # Memberi tahu user bahwa buku berhasil dihapus dari katalog
        except ValueError: # Jika terdapat ValueError
          print("Maaf input salah. Coba lagi ya.\n") # Memberi tahu user untuk menginput ulang

      elif pilih == '4': # Jika user memilih 4
        books = self.show_book() # Memanggil fungsi show_book yaitu menampilkan semua buku yang ada di dalam katalog
        if len(books) > 0: # Jika di dalam katalog terdapat buku
          print("Isi Katalog Buku:") # Menampilkan isi katalog
          for book in books: # For loop untuk menampilkan semua buku yang ada di dalam katalog
            print(f"Judul: {book.title}, Penulis: {book.author}, ID Buku: {book.unique_id}") # Menampilkan detail setiap buku
        else: # Jika katalog kosong (tidak ada buku)
          print("Katalog buku masih kosong.") # Memberi tahu user bahwa katalog masih kosong

      elif pilih == '5': # Jika user memilih 5
        print("Terima kasih sudah menggunakan program katalog buku, sampai jumpa!")
        break # Keluar dari while loop

      else: # Jika user menginput bukan dari salah satu pilihan yang ada di menu
        print("Pilihannya salah. Coba lagi ya.\n") # Memberi tahu user bahwa input nya salah dan coba input ulang
```

```python
catalog = Catalog()  # Membuat kapsul catalog untuk class Catalog
catalog.run_user_interface()  # Menjalankan user interface
```

# Unittest

```python
import unittest

class TestCatalog(unittest.TestCase):
  '''
  Class TestCatalog digunakan untuk mengetest apakah class Catalog
  sudah berjalan dengan baik sesuai yang diharapkan. Pada class
  TestCatalog, akan diuji beberapa fungsi, yaitu fungsi tambah buku
  ke dalam katalog, mencari buku berdasarkan nama penulis, dan menghapus
  buku dari katalog.
  '''

  def test_add_book(self):
    '''
    Fungsi test_add_book adalah fungsi yang digunakan untuk mengetes apakah
    fungsi add_book sudah berjalan sesuai yang diharapkan. Fungsi ini bekerja
    dengan menambahkan buku baru kedalam katalog.
    '''

    catalog = Catalog() # Membuat kapsul catalog dari class Catalog untuk dilakukan testing
    catalog.add_book("Buku Kimia", "Rafa", "001") # Menambahkan buku ke dalam katalog menggunakan fungsi add_book
    self.assertIn("001", catalog.books) # Memeriksa apakah ID buku sudah ada di dalam katalog

  def test_find_book(self):
    '''
    Fungsi test_find_book adalah fungsi yang digunakan untuk mengetes apakah
    fungsi find_book sudah berjalan sesuai yang diharapkan. Fungsi ini bekerja
    dengan mencari buku berdasarkan nama penulis dalam katalog.
    '''

    catalog = Catalog() # Membuat kapsul catalog dari class Catalog untuk dilakukan testing

    # Menambahkan buku-buku ke dalam katalog menggunakan fungsi add_book
    catalog.add_book("Buku Kimia", "Rafa", "001")
    catalog.add_book("Buku Fisika", "Aisya", "002")
    catalog.add_book("Buku Biologi", "Rafa", "003")

    found_book_id = catalog.find_book("Rafa") # Mencari buku berdasarkan nama penulis

    self.assertEqual(len(found_book_id), 2)
    # Memeriksa apakah buku yang ditemukan hanya buku dengan nama penulisnya adalah Rafa
    # Buku dengan nama penulis nya adalah Rafa ada 2 buku
    for book_id in found_book_id: # For loop untuk memeriksa buku yang telah ditemukan
        self.assertEqual(book_id.author, "Rafa")  # Memeriksa apakah buku yang ditemukan nama penulisnya adalah Rafa

  def test_remove_book(self):
    '''
    Fungsi test_remove_book adalah fungsi yang digunakan untuk mengetes apakah
    fungsi remove_book sudah berjalan sesuai yang diharapkan. Fungsi ini bekerja
    dengan menghapus buku dari dalam katalog.
    '''

    catalog = Catalog() # Membuat kapsul catalog dari class Catalog untuk dilakukan testing
    catalog.add_book("Buku Kimia", "Rafa", "001") # Menambahkan buku ke dalam katalog menggunakan fungsi add_book
    catalog.remove_book("001") # Menghapus buku dengan ID 001 dari katalog menggunakan fungsi remove_book
    self.assertNotIn("001", catalog.books) # Memeriksa apakah ID buku sudah tidak ada di dalam katalog

if __name__ == '__main__':
  unittest.main(argv=[''], exit=False)
```
